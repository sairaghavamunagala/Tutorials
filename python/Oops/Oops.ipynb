```python
"""
creation of method and purpose of self.
"""
class Employee:
    
    def __init__(self,first,last,pay):
        self.first=first
        self.last=last
        self.email=first+last+'@'+'gmail.com'
        self.pay=pay

    def full_name(self):
        return f"{self.first} {self.last}"

emp1=Employee('sai','raghava',5000)
emp2=Employee('sai','surya',4000)

print(emp1.email)
print(emp2.email)

print(Employee.full_name(emp1))
print(emp1.full_name())

"""
Why we use self?

For class methods instance of class is passed as argument
if emp1 is passed then, self refers to emp1.
if emp2 is passed then, self refers to emp2.
"""
```

```python
"""
This module explains how to count instance of classes or calculate number of employees.
"""
class Employee:
    raise_amount=1.04
    number_of_employees=0
    def __init__(self,first,last,pay):
        self.first=first
        self.last=last
        self.email=first+last+'@'+'gmail.com'
        self.pay=pay
        Employee.number_of_employees+=1

    def full_name(self):
        return f"{self.first} {self.last}"
    
    def apply_raise(self):
        self.pay=int(self.pay* self.raise_amount) 


print(Employee.number_of_employees)
emp1=Employee('sai','raghava',5000)
emp2=Employee('sai','surya',4000)
print(Employee.number_of_employees)
"""
is instances of the  classes or not.
"""
class Student:
    pass
print(isinstance(emp1,Employee))
print(isinstance(emp1,Student))
"""
Adding class variable and removing class variable.
"""
Employee.sample='sample'
print(Employee.__dict__)
del Employee.sample
print(Employee.__dict__)
```
